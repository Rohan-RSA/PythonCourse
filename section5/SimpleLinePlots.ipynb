Line colors and styles

```python
plt.plot(x, np.sin(x - 0), color='red')        # specify color by name
```

```python
plt.plot(x, np.sin(x - 1), color='blue')            # short color by name
plt.plot(x, np.sin(x - 2), color='g')               # short color by color code(rgbcmyk)
plt.plot(x, np.sin(x - 3), color='0.75')            # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 4), color='#ffdd44')         # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 5), color= (1.0, 0.2, 0.3))  # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 6), color='chartreuse')      # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# for short, you can use the following codes
plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':')
```

```python
plt.plot(x, x + 8, 'g')  # combine color and linestyle
plt.plot(x, x + 9, '--c')   # combine color and linestyle
plt.plot(x, x + 10, '-.k')   # combine color and linestyle
plt.plot(x, x + 11, ':r')   # combine color and linestyle
```
