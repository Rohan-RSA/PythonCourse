```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-v0_8-white')
```

```python
# plt.axes : Subploots by hand

ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])  # [left, bottom, width, height]
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8],    
                    xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                    ylim=(-1.2, 1.2))

x = np.linspace(0, 10)
ax1.plot(np.sin(x), label='sin(x)')
ax2.plot(np.cos(x), label='cos(x)')
```

```python
# plt.subplots : Simple grids of subplots

for i in range (1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, 1)),
             fontsize=18, ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
            fontsize=18, ha='center')
#     ax.set_xticklabels([])
#     ax.set_yticklabels([])
    
```

```python
# plt.gridspec() : more complicated grids of subplots

grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.4)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2])
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 1000).T

fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, wspace=0.2, hspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[] ,sharey=main_ax)
x_hist = fig.add_subplot(grid[:-1, 1:], yticklabels=[] ,sharex=main_ax)

main_ax.plot(x, y, 'ok', markersize=3)

x_hist.hist(x, 40, histtype='stepfilled', color='gray',
            orientation='vertical')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled', color='gray',
            orientation='horizontal')
y_hist.invert_xaxis()
```
